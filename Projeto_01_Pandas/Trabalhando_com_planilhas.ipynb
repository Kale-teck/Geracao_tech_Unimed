```python
import pandas as pd
```

```python
df1 = pd.read_excel('/content/Aracaju.xlsx')
df2 = pd.read_excel('/content/Fortaleza.xlsx')
df3 = pd.read_excel('/content/Natal.xlsx')
df4 = pd.read_excel('/content/Recife.xlsx')
df5 = pd.read_excel('/content/Salvador.xlsx')

#esses caminhos vogam para arquivos locais e não do drive
```

```python
#Juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#amostra aleatoria
df.sample(10)
```

```python
df.dtypes
```

```python
#mudando type de LojaID de int64 para object(string)
df['LojaID'] = df['LojaID'].astype('object')
```

```python
df.isnull().sum()
```

```python
df['Vendas'].mean()
#média de vendas
```

**Criando Colunas**

```python
#criando a coluna receita
df['Receita'] = df['Vendas'].mul(df['Qtde'])
#cria Receita com Vendas * Quantidade
```

```python
df['Receita/Venda'] = df['Receita'] / df['Vendas']
#Cria Receita/Venda com a Recita / vendas
```
